## Exploratory Data Analysis (EDA)

### I. Gender

```python
# Plotting Number of male and female shoppers

sns.countplot(x = "Gender", data = data)
plt.show()
```

```python
# Plot male and female shoppers along with its count value

ax = sns.countplot(x = "Gender", data = data)

#To get the height of the bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# With different color

ax = sns.countplot(x = "Gender", data = data, hue = "Gender", palette = {"F" : "cyan", "M" : "magenta"})

for groups in ax.containers:
    ax.bar_label(groups)

plt.show()
```

```python
# Using HUE attrubute with Gender

ax = sns.countplot(x = "Gender", hue = "Marital_Status", data = data, palette = "Set2")

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
# Using HUE attrubute with Gender but showcasing count vales only for the Married ones

ax = sns.countplot(hue = "Marital_Status", x = "Gender", data = data, palette = "Accent_r")

ax.bar_label(ax.containers[1])

plt.show()
```

```python
# sns.countplot(x = "Gender", data = data, color = "salmon", saturation = 0.9)
# plt.show()
```

```python
# Using different attributes to style the plot

ax = sns.countplot(x = "Gender",
             data = data,
             facecolor = (0, 0, 0, 0),
             linewidth = 5,
             edgecolor = sns.color_palette("BrBG", 2))

for bars in ax.containers:
    ax.bar_label(bars, padding = 3,)

# To push the top most line of the plot little further up
ax.set_ylim(0, ax.get_ylim()[1] * 1.05)
plt.show()
```

```python
# Plot the total amount spend by Male and Female

gen_amt = data.groupby("Gender", as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
gen_amt
```

```python
sns.barplot(x = "Gender", y = "Amount", data = gen_amt )
plt.show()
```

From this plot, we can see that the women has spend more amount during Diwali Sales, as compared to  men.

### II. Age

```python
# To know who shopped themost from which age group
ax = sns.countplot(x = "Age Group", data = data, hue = "Gender", palette = "Accent_r")

for height in ax.containers:
    ax.bar_label(height)
plt.show()
```

<b> From this plot, we can say that the most shoppers are females from the age group 26-35 </b>

```python
# To see people belong to which age group has spent the most money

# Use groupby() for it first
age_amt = data.groupby("Age Group", as_index = False)["Amount"].sum().sort_values("Age Group")
age_amt
```

```python
# Now we will plot the grouped data on a bar plot

sns.barplot(x = "Age Group", y = "Amount", data = age_amt)
plt.show()
```

From this plot we can say that most of the shoppers are of age 26 to 30.

From this we can see that the most amount spent are by the shoppers of Uttar Pradesh, Maharashtra and Karnataka

### III. States

```python
# Using groupby() to aggregate no of orders from each state

st_ord = data.groupby("State", as_index = False)["Orders"].sum().sort_values("Orders", ascending = False)
st_ord
```

```python
# Ploting top 10 states with most no. of orders

plt.figure(figsize=(17, 6))
sns.barplot(x = "State", y = "Orders", hue = "State", data = st_ord.head(10), palette = "gist_stern")
plt.show()
```

```python
# Total Amount spent by shoppers from all the states

# Using groupby()
st_amt = data.groupby("State", as_index = False)["Amount"].sum().sort_values("Amount", ascending = False)
st_amt
```

```python
# Ploting top 10 states with most Amount spent

plt.figure(figsize=(17, 6))
sns.barplot(x = "State", y = "Amount", hue = "State", data = st_amt.head(10), palette = "Set2")
plt.show()
```

From these pllots we can say that the most no. of orders and sales are from Uttar Pradesh, Maharashtra and Karnataka

### IV. Marital Status

```python
data.columns
```

```python
# Ploting shoppers on the basis of marital Status and Gender

ax = sns.countplot(x = "Marital_Status", data = data)
for height in ax.containers:
    ax.bar_label(height)

plt.show()
```

```python
# Grouping data to get amount spend by married and unmarried, male and female

ms_amt = data.groupby(["Marital_Status", "Gender"], as_index = False)["Amount"].sum().sort_values(["Marital_Status", "Gender"])
ms_amt
```

```python
# Ploting shoppers as their marital Status and Gender

sns.barplot(x = "Marital_Status", y ="Amount", hue = "Gender", data = ms_amt, palette = "Set3")
plt.show()
```

From this plot we can see that the most pf the shoppers are Married women and they have spent the most amount

### V. Occupation

```python
data["Occupation"] = data["Occupation"].astype("category")
```

```python
data["Occupation"].cat.categories
```

```python
# Ploting shoppers on the basis of marital Status and Gender

plt.figure(figsize = (20, 8))

ax = sns.countplot(x = "Occupation", data = data)
for height in ax.containers:
    ax.bar_label(height)
plt.show()
```

```python
# Grouping data to get amount spend by shoppers across the occupations

occ_amt = data.groupby(["Occupation"], as_index = False, observed = False)["Amount"].sum().sort_values(["Amount"], ascending = False)
occ_amt
```

```python
# Ploting the Amount spent by shoppers from different occupations

plt.figure(figsize = (20, 8))
sns.barplot(x = "Occupation", y ="Amount", data = occ_amt, order = occ_amt["Occupation"])
plt.show()
```

Here we can see that most of the shopperes are working in IT, Healthcare and Aviation Sector.

### VI. Product Category

```python
data["Product_Category"] = data["Product_Category"].astype("category")
```

```python
data["Product_Category"].cat.categories
```

```python
len(data["Product_Category"].cat.categories)
```

```python
# Ploting shoppers on the basis of Products they bought

plt.figure(figsize = (23, 7))

ax = sns.countplot(x = "Product_Category", data = data)
for height in ax.containers:
    ax.bar_label(height)
plt.show()
```

```python
# Grouping data to get amount spend by shoppers across the Product category

pc_amt = data.groupby(["Product_Category"], as_index = False, observed = False)["Amount"].sum().sort_values(["Amount"], ascending = False).head(10)
pc_amt
```

```python
# Ploting the Amount spent by shoppers from different Product_Category

plt.figure(figsize = (20, 7))
sns.barplot(x = "Product_Category", y ="Amount", data = pc_amt, order = pc_amt["Product_Category"])

plt.show()
```

From above plot we can say that most of the amount was spent on Food, Clothing & Apparels and Electronic Gadgets category.
